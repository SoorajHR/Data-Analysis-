```python
# Top selling courses (and visualisation)
sort=df.sort_values('num_subscribers', ascending=False)
sort
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort the DataFrame by 'num_subscribers' in descending order and select the top 10
top_courses = df.sort_values('num_subscribers', ascending=False).head(10)

# Create the horizontal bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='num_subscribers', y='course_title', data=top_courses)

# Adding labels and title
plt.xlabel('Number of Subscribers', fontsize=12)
plt.ylabel('Course Title', fontsize=12)
plt.title('Top 10 Selling Courses', fontsize=14)

# Improve aesthetics by hiding the top and right spines
plt.gca().spines[['top', 'right']].set_visible(False)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort the DataFrame by 'num_subscribers' in descending order and select the top 10
top_10_courses = df.sort_values('num_subscribers', ascending=False)[['course_title', 'num_subscribers']].head(10)

# Create the horizontal bar chart with 'hue' to avoid the warning
plt.figure(figsize=(10, 6))
sns.barplot(x='num_subscribers', y='course_title', data=top_10_courses, hue='course_title', palette='Blues_r', legend=False)

# Adding labels and title
plt.xlabel('Number of Subscribers', fontsize=12)
plt.ylabel('Course Title', fontsize=12)
plt.title('Top 10 Selling Courses by Number of Subscribers', fontsize=14)

# Improve aesthetics by hiding the top and right spines
plt.gca().spines[['top', 'right']].set_visible(False)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
#PERCENTAGE OF PYTHON GRAPH
```

```python
import matplotlib.pyplot as plt

# Count Python courses
python_courses = df[df['course_title'].str.contains('Python', case=False)].shape[0]

# Count total courses
total_courses = df.shape[0]

# Calculate percentages
python_percent = (python_courses / total_courses) * 100
other_percent = 100 - python_percent

# Create the pie chart
labels = ['Python Courses', 'Other Courses']
sizes = [python_percent, other_percent]
colors = ['#66b3ff', '#99ff99']

plt.pie(sizes, labels= labels, autopct='%1.1f%%', colors=colors, startangle=90,
        textprops={'fontsize': 12, 'weight': 'bold'})  # Increase font size and weight

plt.axis('equal')
plt.title('Percentage of Python Courses', pad=20)  # Add padding to the title


plt.show()
```

```python
#Counting the number of courses published each year
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert the 'published_timestamp' to datetime if it isn't already
df['published_timestamp'] = pd.to_datetime(df['published_timestamp'])

# Extract year from 'published_timestamp'
df['year'] = df['published_timestamp'].dt.year

# Group by year and count the number of courses published each year
courses_per_year = df.groupby('year').size().reset_index(name='num_courses')

# Create a line plot for number of courses published each year
plt.figure(figsize=(10, 6))
sns.lineplot(x='year', y='num_courses', data=courses_per_year, marker='o')

# Add labels and title
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Courses', fontsize=12)
plt.title('Number of Courses Published Each Year', fontsize=14)

# Show plot
plt.tight_layout()
plt.show()
```

```python
#Summing the number of subscribers per year
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert 'published_timestamp' to datetime if not already
df['published_timestamp'] = pd.to_datetime(df['published_timestamp'])

# Extract the year
df['year'] = df['published_timestamp'].dt.year

# Group by year and sum the number of subscribers
subscribers_per_year = df.groupby('year')['num_subscribers'].sum().reset_index()

# Create a line plot for number of subscribers per year
plt.figure(figsize=(10, 6))
sns.lineplot(x='year', y='num_subscribers', data=subscribers_per_year, marker='o')

# Add labels and title
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Subscribers', fontsize=12)
plt.title('Year-wise Number of Subscribers (Demand)', fontsize=14)

# Show plot
plt.tight_layout()
plt.show()
```

```python
#Distribution of Course Levels
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the number of courses for each level
course_levels = df['level'].value_counts().reset_index()
course_levels.columns = ['level', 'num_courses']

# Create a bar plot without using the 'palette' parameter
plt.figure(figsize=(8, 6))
sns.barplot(x='level', y='num_courses', data=course_levels)

# Add labels and title
plt.xlabel('Course Level', fontsize=12)
plt.ylabel('Number of Courses', fontsize=12)
plt.title('Distribution of Courses by Level', fontsize=14)

# Show plot
plt.tight_layout()
plt.show()
```
